```python
import numpy as np
from sklearn_extra.cluster import KMedoids
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
x, y = make_blobs(n_samples=300, centers=3, random_state=42, cluster_std=0.60)
```

```python
plt.scatter(x[:, 0], x[:, 1], s=50, c='gray')
plt.title('Original Data')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
kmedoids = KMedoids(n_clusters=3, random_state=42)
kmedoids.fit(x_train)
```

```python
labels = kmedoids.labels_
medoids = kmedoids.cluster_centers_
```

```python
plt.scatter(x_train[:, 0], x_train[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(medoids[:, 0], medoids[:, 1], c='red', marker='X', s=200, label='Medoids')
plt.title('K-Medoids Clustering (Training Data)')
plt.legend()
plt.show()
```

```python
test= kmedoids.predict(x_test)
```

```python
plt.scatter(x_test[:, 0], x_test[:, 1], c=test, cmap='viridis', s=50)
plt.title('K-Medoids Clustering (Test Data)')
plt.show()
```

```python
from sklearn.metrics import adjusted_rand_score, silhouette_score
```

```python
adjusted_rand_score(y_test, test)
```

```python
silhouette_score(x_test, test)
```
