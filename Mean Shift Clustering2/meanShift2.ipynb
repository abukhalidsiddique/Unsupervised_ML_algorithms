```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
def mean_shift(data, bandwidth, max_iter=10):
    
    shifted_points = np.copy(data)
    fig, ax = plt.subplots(figsize=(10, 6))
    
    ax.scatter(data[:, 0], data[:, 1], c='lightgray', s=50, label="Original Points")
 
    for iteration in range(max_iter):
        new_points = np.copy(shifted_points)
        for i, point in enumerate(shifted_points):
            
            in_bandwidth = data[np.linalg.norm(data - point, axis=1) < bandwidth]
         
            new_point = np.mean(in_bandwidth, axis=0)
            new_points[i] = new_point  # Shift point to this mean location
     
        shifted_points = new_points

        ax.scatter(shifted_points[:, 0], shifted_points[:, 1], alpha=0.3, label=f"Shifted Points (Iter {iteration+1})")

    ax.scatter(shifted_points[:, 0], shifted_points[:, 1], c='blue', s=50, label="Final Shifted Points")
    ax.set_title("Mean Shift Clustering Process")
    ax.legend()
    plt.show()
```

```python
data, _ = make_blobs(n_samples=50, centers=3, cluster_std=1.5, random_state=42)

bandwidth = 3.0  
mean_shift(data, bandwidth)
```
